```python
import numpy as np
import pandas as pd
from keras.utils import to_categorical
```

```python
genomicdata = pd.read_csv(r"chr22.101bp.2000_intervals.JUND.HepG2.tsv",sep="\t")
```

```python
def readGenome(filename):
    genome = ''
    with open(filename, 'r') as f:
        for line in f:
            
            if not line[0] == '>':
                genome += line.rstrip()
    return genome
genome = readGenome(r"lambda_virus.fa") # lambda virüsünün genetik kodunu okumak
genome[:100]
```

```python
data = np.load(r"genomdata.npy") # preprocessed Hepg2 hücre hattı,genomik datası train kısmı
```

```python
labels = np.load(r"genomlabel.npy") #preprocessed genomik data label kısmı
labels = to_categorical(labels)
```

```python
data.shape
```

```python
labels.shape
```

```python
from keras.models import Sequential                   # Model mimarisi 1D CNN Kullandığımız datada 2000 DNA parçası 
                                                    #bulunmakta daha fazla veri ve kromatin erişilebilirliği ile başarı arttırılabilir
from keras.layers import Conv1D, Flatten, Dense,Dropout
from keras.optimizers import Adam

model = Sequential()
model.add(Conv1D(30, 25, input_shape=(101, 4),padding='same'))
model.add(Conv1D(30, 25,padding='same'))
model.add(Conv1D(30, 25,padding='same'))
model.add(Conv1D(30, 25,padding='same'))
model.add(Conv1D(30, 25,padding='same'))
model.add(Flatten())
model.add(Dense(256))
model.add(Dense(512))
model.add(Dropout(0.5))
model.add(Dense(2, activation='sigmoid'))

```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(lr=.0001), metrics=['accuracy'])
model.fit(data[:1900],labels[:1900],epochs=700,batch_size=500,validation_data=(data[1900:2000],labels[1900:2000]),verbose=2)
```
